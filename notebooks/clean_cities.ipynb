```python
import pandas as pd 
```

```python
df = pd.read_csv("../tools/communes-francaises.csv")
df.head()
```

```python
df.info()
```

```python
def clean_city_name(city_name):
    allowed_chars = set(" -'")  # espaces, tirets, apostrophes
    cleaned = []

    for ch in city_name:
        if ch.isalpha() or ch in allowed_chars:
            cleaned.append(ch)

    cleaned_name = "".join(cleaned)
    cleaned_name = " ".join(cleaned_name.split())  # retirer espaces multiples
    return cleaned_name

df['cleaned_city_name'] = df['Commune'].apply(clean_city_name)
df.head()
```

```python
df.drop_duplicates(subset=['cleaned_city_name'], inplace=True)
df.to_csv("../tools/cleaned_communes_francaises.csv", index=False)
```
